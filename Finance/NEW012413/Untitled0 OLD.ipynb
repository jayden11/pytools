```python
from __future__ import division
import pandas as pd
import numpy as np
from pandas.io.parsers import ExcelWriter

pd.set_printoptions(max_rows=1000, max_columns=20)

combined_file = pd.io.parsers.ExcelFile('Combined_Volatility.xlsx')
df = combined_file.parse('All')
sorts = df.set_index('Ticker')  # Only has lines with Annual Returns >= 0.7 (7%)
only_divs = sorts[sorts['Ex. Div Date'] != 'None']
```

```python
stable = only_divs[only_divs.Volatility <= 0.08]
```

```python
close = stable[np.abs(stable['Gain/Loss Exercise']) <= 1]
```

```python
close['(Div/Share)/P'] = close['Div/share'] / close['Stock Price']
```

```python
signif = close[['Name', 'Industry', 'Stock Price', 'Strike', 'Option Price', 'Exp. Date', 'Option Volume', 
                'Ex. Div Date', 'Div/share', 'Annual Return', 'Volatility', '(Div/Share)/P']]
```

```python
signif.sort(columns='Annual Return', ascending=0)
```

```python
signif[signif['Industry'] == 'Precious Metals'].sort(columns='Annual Return', ascending=0)
```
